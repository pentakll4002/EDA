# EDA and Feature Engineering Flight Price Prediction

#### 1. Airline: The name of the airline company is stored in the airline column. It is a categorical feature having different airlines.
#### 2. Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
#### 3. Duration: A continuous variable that stores the total duration of the flight in minutes.
#### 4. Departure Time: This is a derived categorical feature obtained by grouping time periods into bins. It stores information about the departure time in discrete time labels.
#### 5. Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
#### 6. Arrival Time: A categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
#### 7. Destination City: The city where the flight will land. It is a categorical feature having 6 unique cities.
#### 8. Date: A continuous variable that stores the date of the flight. It has two distinct values: Business and Economy.
#### 9. Duration: A continuous variable that displays the overall amount of time it takes to travel between cities in days.
#### 10. Ticket Price: Target variable stores information of the ticket price.

```python
### Feature Engineering
# Lấy ngày ra thì (0)
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype('int')
df['Month'] = df['Month'].astype('int')
df['Year'] = df['Year'].astype('int')
```

```python
df.info()
```

```python
### Drop Date Of Journey
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
### Chỉ lấy phần đầu tiên trong cột dữ liệu
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype('int')
df['Arrival_min'] = df['Arrival_min'].astype('int')
```

```python
### Drop Arrival_Time
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour'] = df['Departure_hour'].astype('int')
df['Departure_min'] = df['Departure_min'].astype('int')
```

```python
df.info()
```

```python
### Drop Dep_Time
df.drop('Dep_Time', axis=1, inplace=True)
```
